# Fitting compartmental models to generation time distributions

```julia
using OrdinaryDiffEq,Plots,Distributions,SparseArrays
```

```julia
function create_rate_matrix(p)
    σ₁, σ₂, γ,β = p
    Q = spzeros(5,5)
    Q[2,1] = 2*σ₁ #E1->E2
    Q[3,2] = 2*σ₁ #E2->P
    Q[4,3] = σ₂ #P->I
    Q[5,4] = γ #I->R
    for k = 1:4
        Q[k,k] = -sum(Q[:,k])
    end
    return Q
end
```

```julia
p = [1/2,1/1,1/3,2/4]
Q = create_rate_matrix(p)
L = DiffEqArrayOperator(Q) 
```

```julia
# function f(du,u,p,t)
#     du .= Q*u
# end
u0 = [1.,0.,0.,0.,0.]
tspan = (0,80.)
ff = ODEFunction(L,syms = [:prob_E1,:prob_E2,:prob_P,:prob_I,:prob_R])
prob = ODEProblem(ff,u0,tspan,p)
@time sol = solve(prob,Tsit5())




plot(sol)
```

```julia


function generation_time_from_sol(sol)
    p = sol.prob.p
    σ₁, σ₂, γ,β = p
    R₀ = β*(0.1*(1/σ₂) + (1/γ) )
    return t -> sum(sol(t).*[0.,0.,0.1*β,β,0])/R₀
end
```

```julia
w_t = generation_time_from_sol(sol)
plt = plot(w_t,0.,90.,lab = "generation time distribution")
d_gen_time = Weibull(2.826,5.665)
fitted_gen_time = [pdf(d_gen_time,t) for t in 0:0.1:20]
plot!(plt,0:0.1:20.,fitted_gen_time,lab = "Ferreti et al fit")
```

```julia
function rel_entropy(w_t)
    p_w = [w_t(t) for t in 0.:0.1:20. ]
    return  0.1*sum([p_w.*(log.(p_w) - log.(fitted_gen_time))])
end
    
```

```julia
rel_entropy(w_t)
```
