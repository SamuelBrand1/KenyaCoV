# KenyaCoV vs deterministic model --- test Stochastic vs ODE limit

The purpose of this short note is to demonstrate that the stochastic Kenya ncov model behaves as expected compared to an ODE model. We initiate the model by including 20-25 year old 10 pre-symptomatic infecteds amongst the population in each area. This locks in epidemic growth from time zero and effectively eliminates stochastic effects, e.g. stochastic fade-out, stochastic invasion. We also have a brief look at the effect of time-step size on the typical duration of infection.

In this regime, the dynamics of the stochastic model **should** closely match the dynamics of the ODE model with the same underlying transmission structure because the population size is large in each county.

For simplicity there is no treatment and no external imports. Pre-symptomatics and asymptomatics are assumed to be as infectious as symptomatics. Baseline parameter estimates R₀ = 2.5.

Having defined the basic parameters we can solve and plot two different solutions:
* KenyaCoV with timestep dt = 0.25
* KenyaCoV solved as the related deterministic model

```julia
push!(LOAD_PATH, "/Users/Sam/GitHub/KenyaCoV/src")
push!(LOAD_PATH, "/Users/Sam/GitHub/KenyaCoV/data")

using KenyaCoV, DifferentialEquations, Plots
using LinearAlgebra
```

```julia
# Load data for model --- and create an initially susceptible population
u0,P,transport_matrix = KenyaCoV.model_ingredients_from_data("./../data/data_for_age_structuredmodel.jld2",
                                                             "./../data/flight_numbers.csv",
                                                             "./../data/projected_global_prevelance.csv");
# #Change some parameters away from default

for (i,p) in enumerate(P.global_prev)
    P.global_prev[i] = 0.
end
P.ϵ = 1.
P.χ = ones(17)

#Rescale by maximum eigenvalue of the age mixing matrix rescaled by susceptibility

sus_matrix = repeat(P.χ,1,17)
R_A = P.ϵ*((1/P.σ₂) + (1/P.γ) )
R_M = (P.ϵ/P.σ₂) + (P.ϵ_D/P.γ)
R_V = (P.ϵ/P.σ₂) + (P.ϵ_V/P.τ)
R_vector = [(1-P.rel_detection_rate[a])*R_A + P.rel_detection_rate[a]*(1-P.hₐ[a])*R_M + P.rel_detection_rate[a]*P.hₐ[a]*R_V for a = 1:17]
inf_matrix = repeat(R_vector',17,1)

eigs, = eigen(sus_matrix.*P.M.*inf_matrix)
max_eigval = Real(eigs[end])


P.β = 2.5/max_eigval# Choosing the transmission rate that matches R₀ = 2.5
P.dt = 0.1; #<---- KenyaCoV is a discrete time simulation, timestep here is 1/10th of a day
# #Add 10 Pre-symptomatic infecteds in 20-24 year old age category in every county
for i = 1:KenyaCoV.n_wa
    u0[i,5,3] += 10
end
```

```julia
#Solve two different models
P.dt = 0.25
prob = KenyaCoV.create_KenyaCoV_non_neg_prob(u0,(0.,365.),P)
sol = solve(prob,FunctionMap(),dt = P.dt)

prob_ode = KenyaCoV.create_KenyaCoV_ode_prob(u0,(0.,365.),P)
sol_ode = solve(prob_ode,Tsit5());
```

```julia
PreS = [sum(u[:,:,3]) for u in sol.u ]
PreS_ode = [sum(sol_ode(t)[:,:,3]) for t in sol.t] 

plt = plot(sol.t,PreS .+ 1,lw = 3, lab = "KenyaCoV-stochastic, dt = $(P.dt)", xlims = (0.,240.), xlabel = "time (days)", 
                    ylabel = "Pre-symptomatics + 1",
                    yscale = :log10)
plot!(plt,sol.t,PreS_ode .+ 1,lw = 3, ls = :dash,color = :black, lab = "KenyaCoV-ODE")
```

```julia
cum_PA = [sum(u[:,:,9]) for u in sol.u ]
cum_PA_ode = [sum(sol_ode(t)[:,:,9]) for t in sol.t] 

plt = plot(sol.t,cum_PA .+ 1,lw = 3, lab = "KenyaCoV-stochastic, dt = $(P.dt)", xlims = (0.,240.), xlabel = "time (days)", 
                    ylabel = "Cum. A infs + 1",
                    yscale = :log10)
plot!(plt,sol.t,cum_PA_ode .+ 1,lw = 3, ls = :dash,color = :black, lab = "KenyaCoV-ODE")
```

```julia
cum_PM = [sum(u[:,:,10]) for u in sol.u ]
cum_PM_ode = [sum(sol_ode(t)[:,:,10]) for t in sol.t] 

plt = plot(sol.t,cum_PM .+ 1,lw = 3, lab = "KenyaCoV-stochastic, dt = $(P.dt)", xlims = (0.,240.), xlabel = "time (days)", 
                    ylabel = "Cum. mild infs + 1",
                    yscale = :log10)
plot!(plt,sol.t,cum_PM_ode .+ 1,lw = 3, ls = :dash,color = :black, lab = "KenyaCoV-ODE")
```

```julia
cum_PV = [sum(u[:,:,10]) for u in sol.u ]
cum_PV_ode = [sum(sol_ode(t)[:,:,10]) for t in sol.t] 

plt = plot(sol.t,cum_PV .+ 1,lw = 3, lab = "KenyaCoV-stochastic, dt = $(P.dt)", xlims = (0.,240.), xlabel = "time (days)", 
                    ylabel = "Cum. severe infs + 1",
                    yscale = :log10)
plot!(plt,sol.t,cum_PV_ode .+ 1,lw = 3, ls = :dash,color = :black, lab = "KenyaCoV-ODE")
```
